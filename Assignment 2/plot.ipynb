```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


sns.set()
```

```python
df1 = pd.read_csv("adam.csv")
df1 = df1.drop(columns="Epoch")
sns.lineplot(data=df1)
plt.title("Training error for Adam")
plt.xlabel("Epoch")
plt.ylabel("Cross entropy loss")
plt.savefig('adam', dpi=600)
```

```python
df2 = pd.read_csv("sgd.csv")
df2 = df2.drop(columns="Epoch")
sns.lineplot(data=df2)
plt.title("Training error for SGD")
plt.xlabel("Epoch")
plt.ylabel("Cross entropy loss")
plt.savefig('sgd', dpi=600)
```

```python
df3 = pd.read_csv("momen.csv")
df3 = df3.drop(columns="Epoch")
sns.lineplot(data=df3)
plt.title("Training error for SGD with momentum")
plt.xlabel("Epoch")
plt.ylabel("Cross entropy loss")
plt.legend(title="Momentum")
plt.savefig('momen', dpi=600)
```

```python
df4 = pd.read_csv("cnn.csv")
df4 = df4.drop(columns="Epoch")
sns.lineplot(data=df4)
plt.title("Training error for original and modified CNN with Adam")
plt.xlabel("Epoch")
plt.ylabel("Cross entropy loss")
plt.savefig('cnn', dpi=600)
```
